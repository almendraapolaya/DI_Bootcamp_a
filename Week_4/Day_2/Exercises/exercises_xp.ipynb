Exercise 1 : Matrix Operations
===
Instructions:

Create a 3x3 matrix and perform the following operations:

Calculate the determinant.
Find the inverse of the matrix.

```python
import numpy as np
import pandas as pd
```

```python
matrix = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [5, 6, 2]
])

det_matrix = np.linalg.det(matrix).round()

inverse_matrix = np.linalg.inv(matrix)

print(f"Determinant: {det_matrix}")
print(f"Inverse matrix: {inverse_matrix}")
```

Exercise 6 : Basic Hypothesis Testing
===
Instructions

Create a sample dataset to test the effectiveness of a new training program on employee productivity.

Given a dataset, formulate a simple hypothesis and test it using basic statistical functions in NumPy.

```python
np.random.seed(110)

productivity_before = np.random.normal(loc=50, scale=10, size=30)

productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

differences = productivity_after - productivity_before

mean_improvement = np.mean(differences)
std_improvement = np.std(differences)

print(f"Average Productivity Increase: {mean_improvement:.2f}")
print(f"Consistency of Increase (Std Dev): {std_improvement:.2f}")

if mean_improvement > 0:
    print("Conclusion: The training program was effective. Productivity increased!")
else:
    print("Conclusion: The training program was not effective. No increase detected.")
```

The statistical analysis of the productivity scores reveals a positive trend following the implementation of the training program. By calculating the difference between the "before" and "after" states, we observed a mean improvement ($+5.0$ on average), suggesting that the program successfully enhanced employee output. Furthermore, the standard deviation of these differences allows us to measure the consistency of this growth; a lower deviation indicates that the training benefited the entire group uniformly, while a higher one suggests varied individual responses. In summary, because the average change is greater than zero, we reject the null hypothesis (that the training had no effect) and conclude that the program was an effective investment for increasing productivity.

Exercise 8 : Time Series Data Manipulation
===
Instructions

Generate time series data for the year 2023. Demonstrate slicing for the following intervals:

January to March
April to June
July to September
October to December
Generate a time series data for a specific period and demonstrate how to slice this data for different intervals.

```python
import pandas as pd
import numpy as np

np.random.seed(100)

dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

data = np.random.randint(50, 200, size=len(dates))

ts_df = pd.DataFrame(data, index=dates, columns=['Activity'])

print("First 5 rows of our Time Series:")
print(ts_df.head())
```

Slicing, verifying and Aggregating:

```python
# Q1: January to March
q1 = ts_df['2023-01-01':'2023-03-31']

# Q2: April to June
q2 = ts_df['2023-04-01':'2023-06-30']

# Q3: July to September
q3 = ts_df['2023-07-01':'2023-09-30']

# Q4: October to December
q4 = ts_df['2023-10-01':'2023-12-31']

print(f"Q1 Average Activity: {q1.mean().values[0]:.2f}")
print(f"Q2 Average Activity: {q2.mean().values[0]:.2f}")
print(f"Q3 Average Activity: {q3.mean().values[0]:.2f}")
print(f"Q4 Average Activity: {q4.mean().values[0]:.2f}")
```

Exercise 10 : Basic Visualization
===
Instructions:

Use Matplotlib to visualize a simple dataset created with NumPy (e.g., a line graph of random numbers).

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed()

days = np.arange(1, 11)
streams = np.random.randint(500, 1500, size=10)
```

```python
plt.plot(days, streams, marker='o', color='b', linestyle='--')

plt.title("Daily Song Streams (First 10 Days)")
plt.xlabel("Day Number")
plt.ylabel("Number of Streams")

plt.grid(True)

plt.show()
```

The line graph illustrates a highly volatile initial period for the song's performance, peaking on Day 4 with nearly 1,500 streams. Following this peak, there is a noticeable downward trend (or "cooldown" period), reaching a low point on Day 8. However, the slight uptick on Day 9 suggests a potential stabilization or a small resurgence in interest. For a data analyst, this visual representation is crucial because it reveals that while the song has strong "viral" potential (seen in the early spikes), the current marketing or listener retention may need adjustment to maintain consistent growth and avoid the sharp decline seen in the latter half of the week.
