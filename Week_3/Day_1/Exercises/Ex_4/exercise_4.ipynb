```python
import pandas as pd
```

```python
df_iris = pd.read_csv('iris.csv')

print("Dataset loaded successfully!")
```

First rows of Iris dataset:

```python
df_iris.head()
```

Quantitative and Qualitative columns:

```python
df_iris['Id'] = df_iris['Id'].astype(str)

quant_iris = df_iris.select_dtypes(include=['number']).columns.tolist()
qual_iris = df_iris.select_dtypes(exclude=['number']).columns.tolist()

print(f"Quantitative {quant_iris}")
print(f"\nQualitative: {qual_iris}")
```

Quantitative:
The columns SepalLengthCm, SepalWidthCm, PetalLengthCm, and PetalWidthCm are categorized as Quantitative because they represent physical measurements. Calculating an average for these columns is mathematically meaningful because it helps describe the "typical" size of a flower in the dataset.

Qualitative:
The columns Species and Id are categorized as Qualitative because they represent nominal identifiers.

-Species labels the group a flower belongs to (e.g., Iris-setosa).

-Id is a unique numeric label used for identification rather than measurement.

These columns are qualitative because performing arithmetic on them—such as adding two IDs together or averaging the species names—would result in values that have no real-world meaning.
