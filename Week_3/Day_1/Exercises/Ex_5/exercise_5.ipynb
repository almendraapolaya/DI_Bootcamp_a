Group Comparison and Trend Analysis:

Columns: Sex, Age Group, and Avg hrs per day sleeping.

I will focus specifically on the 'Men' and 'Women' categories to observe the differences in their sleep patterns

Explanation: I chose Sex because ".unique()" showed that the data is divided into 'Men' and 'Women,' which is perfect for a Group Comparison to see who sleeps more.

I chose Age Group because it allows for a Trend Analysis to see if sleep changes as people move from '15-24 years' to '65 years and over.'

I chose Avg hrs per day sleeping because my .describe() results showed a wide range of sleep times (from 7.7 to 10.6 hours), and I want to see how that variety is distributed among the different groups.

```python

women_sleep = df_american_sleep[df_american_sleep['Sex'] == 'Women']
print("Average sleep for Women:", women_sleep['Avg hrs per day sleeping'].mean())

men_sleep = df_american_sleep[df_american_sleep['Sex'] == 'Men']
print("\nAverage sleep for Men:", men_sleep['Avg hrs per day sleeping'].mean())
```
