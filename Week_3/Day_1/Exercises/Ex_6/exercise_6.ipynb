Exercise 6: Identifying Data Types

Below are various data sources. Identify whether each one is an example of structured or unstructured data.

- A company’s financial reports stored in an Excel file.
- Photographs uploaded to a social media platform.
- A collection of news articles on a website.
- Inventory data in a relational database.
- Recorded interviews from a market research study.

A company’s financial reports in an Excel file:
________________________________________________
 Classification: Structured Data 

 Financial reports in Excel are organized into rows and columns with clear headers (like "Revenue" or "Date"). Because each piece of data has a fixed place and a specific format (usually numbers or dates), a computer can easily search and calculate it.
 _______________________________________________

Photographs uploaded to a social media platform:
___________________________________________________
Classification: Unstructured Data 

Photos do not have a pre-defined row and column structure. While they contain a lot of information (colors, faces, locations), the data is "random" to a computer until it uses special AI to "read" what is inside the image.
_________________________________________________

A collection of news articles on a website:
_______________________________________________
Classification: Unstructured Data 

Even though an article has a title and a body, the actual content is "free-form" text. Because the length, vocabulary, and sentence structure change every time, it doesn't fit into a tidy spreadsheet.
________________________________________________

Inventory data in a relational database:
__________________________________________
Classification: Structured Data 

 Relational databases (like SQL) are the definition of structured data. They use strict tables where every "Product ID" or "Quantity" must follow a specific rule, making it very easy to manage and analyze.
 ___________________________________________

Recorded interviews from a market research study:

Classification: Unstructured Data 

Audio files are similar to photos. The data consists of sounds, tones, and pauses. There is no internal "table" for the information, so it requires transcription or specialized software to turn it into something a computer can analyze as "data."
