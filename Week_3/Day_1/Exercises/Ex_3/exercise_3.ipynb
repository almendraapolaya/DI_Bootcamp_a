Exercise 3: Identifyng Data Types

For the datasets from the previous exercise, categorize each column of it as either quantitative or qualitative and explain your reasoning.

```python
import pandas as pd

df_sleep = pd.read_csv('../Ex_2/sleep.csv')
df_mental = pd.read_csv('../Ex_2/mental_health.csv')
df_credit = pd.read_csv('../Ex_2/credit.csv')

print("Datasets loaded successfully from the Ex_2 folder!")
```

```python
df_sleep.info()
```

```python

print("--- Sleep Dataset Types ---")
print(df_sleep.dtypes)
```

```python
quant_sleep = df_sleep.select_dtypes(include=['number']).columns.tolist()
qual_sleep = df_sleep.select_dtypes(exclude=['number']).columns.tolist()

print(f"Quantitative = {quant_sleep}")
print(f"\nQualitative = {qual_sleep}")
```

```python
print("--- Mental Health Data Types ---")
print(df_mental.dtypes)
```

```python
df_mental['index'] = df_mental['index'].astype(str)

quant_mental = df_mental.select_dtypes(include=['number']).columns.tolist()

qual_mental = df_mental.select_dtypes(exclude=['number']).columns.tolist()

print(f"Quantitative: {quant_mental}")
print(f"\nQualitative: {qual_mental}")
```

```python
df_credit.info()
```

```python
df_credit.head()
```

```python

cols_to_fix = ['ZipCode', 'PriorDefault', 'Employed'] 

for col in cols_to_fix:
    if col in df_credit.columns:
        df_credit[col] = df_credit[col].astype(str)

quant_credit = df_credit.select_dtypes(include=['number']).columns.tolist()
qual_credit = df_credit.select_dtypes(exclude=['number']).columns.tolist()

print(f"Quantitative: {quant_credit}")
print(f"Qualitative: {qual_credit}")
```

Reasons:

Quantitative: These columns represent measurable magnitudes or counts, such as Age, Sleep Duration, and Income. They are categorized as quantitative because their numerical values exist on a scale where mathematical operations—like calculating an average or a sum provide meaningful insights into the data's physical or financial properties.

Qualitative: These columns represent descriptive labels or characteristics, such as Gender, Country, Zip Code, and Binary (0/1) flags. They are categorized as qualitative because they function as nominal identifiers to group observations. Even when they contain digits, performing arithmetic on them (like adding two Zip Codes) is mathematically meaningless because they identify "what kind" rather than "how much.
