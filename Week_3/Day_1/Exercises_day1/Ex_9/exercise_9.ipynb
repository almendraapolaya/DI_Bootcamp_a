Exercise 9 : Export a dataframe to excel format 
and JSON format
====
- Create a simple dataframe.

- Export the dataframe to an excel file.

- Export the dataframe to a JSON file.

```python
%pip install openpyxl
```

```python
import pandas as pd
```

```python

data = {
    'Name': ['Alme', 'Itay', 'Lancelot'],
    'Score': [95, 88, 100]
}

df_mydata = pd.DataFrame(data)
```

Creating an excel:

```python
df_mydata.to_excel('my_exercise_data.xlsx', index=False)

print("Excel file created!")
```

Creating a Json file:

```python

df_mydata.to_json('my_exercise_data.json')

print("JSON file created!")
```
