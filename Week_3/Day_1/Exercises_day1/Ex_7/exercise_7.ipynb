Exercise 7: Transformation Exercise
====
_______________________________________
For each of the following unstructured data sources, propose a method to convert it into structured data. Explain your reasoning.

- A series of blog posts about travel experiences.
- Audio recordings of customer service calls.
- Handwritten notes from a brainstorming session.
- A video tutorial on cooking.


Blog posts about travel experiences:
____________________________________________
Transformation Method: Use Sentiment Analysis and Keyword Extraction.

I can create a spreadsheet with columns like Destination, Rating (1-5), and Sentiment (Positive/Negative). By scanning the text for specific city names and "mood" words (like "amazing" vs "crowded"), you turn a long story into a row of data that you can filter and count.
____________________________________

Audio recordings of customer service calls:
__________________________________________
Transformation Method: Speech-to-Text Transcription followed by Categorization.

First, convert the audio into text. Then, create a table with columns like Call Duration, Reason for Call (e.g., Refund, Technical, Billing), and Resolution (Yes/No). This turns a sound file into a list of searchable issues that a manager can analyze.
____________________________________________

Handwritten notes from a brainstorming session:
_____________________________________________
Transformation Method: OCR (Optical Character Recognition) and Thematic Coding.

Use a scanner or app to turn the handwriting into digital text. Then, you can categorize the ideas into a table with columns like Idea Owner, Category (e.g., Marketing, Product, Design), and Priority Level. This makes a messy page of notes easy to sort in Excel.
______________________________________________

A video tutorial on cooking:
_________________________________
Transformation Method: Object Detection and Step-by-Step Logging.

You can watch the video and extract the data into a spreadsheet with columns like Ingredient Name, Quantity, Cooking Technique (e.g., Frying, Boiling), and Time Spent. This turns a 10-minute visual experience into a structured recipe or a data list of movements and tools used.
