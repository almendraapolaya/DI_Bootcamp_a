**Why data visualization is important in data analysis?**


Data visualization is essential in data analysis because it acts as the primary "lens" through which complex, raw information becomes understandable to the human brain. While raw data consists of abstract numbers, visualization leverages our innate visual processing power—which is 60,000 times faster than text processing—to make insights immediately obvious


Data visualization is the bridge between raw numbers and human understanding. It is critical because:


**Humans are Visual**: Our brains process images 60,000 times faster than text. Seeing a line graph reveals a trend instantly, whereas a spreadsheet requires minutes of mental calculation.


**Pattern Recognition**: It exposes hidden relationships, seasonal cycles, and clusters that are mathematically present but visually invisible in a table.


**Error Detection**: It’s the fastest way to spot outliers or data entry errors—a single dot far away from a cluster screams for attention.


**Effective Communication**: It strips away technical jargon, allowing analysts to tell a clear story that stakeholders can actually act on.


**Faster Decision-Making**: By condensing thousands of data points into one view, it provides the "big picture" needed to make high-stakes moves quickly.

**Describe the purpose of a line graph in data visualization**

A line graph (or line chart) is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.

The Purpose:
The primary purpose of a line graph is to track changes over periods of time. While other charts might show how parts make up a whole, the line graph is the king of continuity.

Showing Trends: It tells you if something is increasing, decreasing, or staying the same over days, months, or years.

Identifying Volatility: It shows how "bumpy" the data is. Are there wild swings (high volatility) or is the growth steady and smooth?

Comparison: You can easily layer multiple lines on one graph to see how two different things (like the stock price of Apple vs. Google) behaved over the same timeframe.

Exercise 2: Creating a Line Plot for Temperature Variation
===
Objective: Create a simple line plot using Matplotlib that represents temperature variations over a week.


Tasks:
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]).
Label the x-axis as “Day” and the y-axis as “Temperature (°F)”.
Add a title to the plot.
Display the plot.

```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.plot(days, temperatures)
plt.xlabel("Day")
plt.ylabel("Temperature (°F)")
plt.title("Weekly Temperature Variations")
plt.show()
```

This plot demonstrates that while the temperature fluctuated, there was a clear peak mid-week (Friday), a insight that is much easier to "see" on a graph than to "read" in a list.

Exercise 3: Visualizing Monthly Sales with a Bar Chart
===
Objective: Generate a bar chart using Matplotlib to visualize monthly sales data for a retail store.
Tasks:
- Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]).
- Label the x-axis as “Month” and the y-axis as “Sales Amount ($)”.
Add a title to the bar chart.
- Display the plot.

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [5000, 5500, 6200, 7000, 7500]

plt.bar(months, sales)

plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")

plt.title("Monthly Sales Performance - Retail Store")

plt.show()
```

With Bar chart, we can quickly see that May was the highest-performing month, showing a steady growth trend from January.

Exercise 5: Comparing Anxiety Levels Across Different Genders
===
Objective: Use a bar plot to compare the proportion of students experiencing anxiety across different genders.
Dataset Overview: The dataset includes columns: ‘Do you have Anxiety?’ and ‘Choose your gender’.
Tasks:
- Import necessary libraries.
- Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.
- Customize the plot with an appropriate color palette and add a title.
- Display the plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Student_Mental_health.csv')

df['Anxiety_Numeric'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})
```

```python
sns.barplot(data=df, x="Choose your gender", y="Anxiety_Numeric", palette="pastel")

plt.title("Proportion of Students with Anxiety by Gender")

plt.ylabel("Proportion (1.0 = 100%)")
plt.show()
```

The objective of this exercise was to determine if a correlation exists between gender and anxiety levels in this dataset. By using a Bar Plot, we successfully visualized the rate of anxiety rather than just the raw count. This allows for a fair comparison even if there are more females than males in the survey.

Exercise 6: Exploring the Relationship Between Age and Panic Attacks
===
Objective: Create a scatter plot to explore the relationship between students’ age and the occurrence of panic attacks.
Dataset Overview: The dataset includes columns: ‘Age’ and ‘Do you have Panic Attacks?’.
Tasks:
- Import necessary libraries.
- Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
- Use Seaborn’s scatterplot to create a scatter plot with ‘Age’ on the x-axis and numeric panic attack responses on the y-axis.
- Customize the plot to improve readability by adding labels, a title, and adjusting point styles.
- Display the plot.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Student_Mental_health.csv')

df['Panic_Numeric'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

sns.scatterplot(data=df, x='Age', y='Panic_Numeric', hue='Do you have Panic attack?', style='Do you have Panic attack?', s=100)

plt.title('Relationship Between Age and Panic Attacks')
plt.xlabel('Age of Students')
plt.ylabel('Panic Attack (0 = No, 1 = Yes)')
plt.yticks([0, 1]) 

plt.show()
```

Looking at the scatter plot, we notice that points are scattered across almost all ages for both "Yes" and "No." This suggests that in this specific group of students, panic attacks aren't strictly limited to older or younger students; they occur across the entire age range (18-24). Without a clear diagonal line or "clustering" at one end, we might conclude there isn't a strong linear relationship between age and the occurrence of panic attacks for this sample.
