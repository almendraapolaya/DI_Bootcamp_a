===

- Download and import the Data Science Job Salary dataset.
- Normalize the ‘salary’ column using Min-Max normalization which scales all salary values between 0 and 1.
- Implement dimensionality reduction like Principal Component Analysis (PCA) or t-SNE to reduce the number of features (columns) in the dataset.
- Group the dataset by the ‘experience_level’ column and calculate the average and median salary for each experience level (e.g., Junior, Mid-level, Senior).




```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

Reading the cvs file:

```python
df = pd.read_csv('ds_salaries.csv')
```

```python
df.head()
```

Normalizing the "salary" column: 

```python
scaler = MinMaxScaler()

df['salary_normalized'] = scaler.fit_transform(df[['salary']])
```

PCA:

```python
from sklearn.decomposition import PCA
```

```python
numerical_df = df.select_dtypes(include=['number'])
```

```python
pca = PCA(n_components=2)
pca_results =  pca.fit_transform(numerical_df)

df_pca = pd.DataFrame(data=pca_results, columns=['PC1', 'PC2'])
```

Grouping and Aggregating:

```python
grouped_data = df.groupby('experience_level')

salary_stats = grouped_data['salary'].agg(['mean', 'median'])
```

```python
print(salary_stats)
```

```python
salary_stats = grouped_data['salary'].agg(['mean', 'median']).round(2)

print(salary_stats)
```


### Key Findings from Salary Analysis

* **General Trend:** There is a consistent increase in pay as employees move from Entry-level to Senior roles.


* **The Executive Anomaly:** While Executives have the highest average pay (\$76k), their "middle" pay (median) is much lower at \$46k.


* **What this means:** This huge gap suggests that the Executive group is "top-heavy." A few people are making a lot of money, which pulls the average up, but most people in that category are actually earning closer to \$46k.


* **Conclusion:** Senior-level roles appear to have a more stable and higher "typical" salary than Executive roles in this specific dataset.
