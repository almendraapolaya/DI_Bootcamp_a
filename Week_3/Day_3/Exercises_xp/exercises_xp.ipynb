Exercise 2: Descriptive Statistics
===
Task: Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.


Sample Dataset:
**data = [12, 15, 13, 12, 18, 20, 22, 21]**

```python
from scipy import stats
import numpy as np
```

Calculating Mean and Median:

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_value = np.mean(data)
median_value = np.median(data)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
```

Calculating Variance and Standard Deviation:

```python
variance_value = stats.tvar(data)
std_dev_value = stats.tstd(data)

print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")
```

Exercise 4: T-Test Application
=====
Task: Perform a T-test on two sets of randomly generated data.

```python
import numpy as np
from scipy import stats

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)
```

Performing the T-test:

```python
t_stat, p_value = stats.ttest_ind(data1, data2)
```

Results:

```python
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

In this exercise, I used a T-test to compare two groups of data. Even though both groups looked similar, their averages were different (50 vs. 60). The T-test gave me a P-value, which is like a "luck-meter."

Because the P-value was very small (less than 0.05), it told me that the difference between the two groups was real and not just a coincidence.

Exercise 5: Linear Regression Analysis
===

Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.


Dataset:
house_sizes: [50, 70, 80, 100, 120] (in square meters)
house_prices: [150,000, 200,000, 210,000, 250,000, 280,000] (in currency units)


**Questions**:
- What is the slope and intercept of the regression line?
- Predict the price of a house that is 90 square meters.
- Interpret the meaning of the slope in the context of housing prices.

```python
import numpy as np
from scipy import stats

house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]
```

Regression:

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

**What is the slope and intercept?**

**Slope**: Approximately 1828.77

**Intercept**: Approximately 64383.56

Predicting:

```python
predicted_price = (slope * 90) + intercept
print(f"Predicted price for 90 sqm: {predicted_price}")
```

In exercise 5, I used Linear Regression to find the mathematical relationship between house size and price. By calculating the Slope, I discovered how much each additional square meter adds to a home's value. This allowed me to create a "prediction formula" to estimate the price of houses that weren't even in my original list.

Exercise 6: Understanding ANOVA
===

Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.


Dataset:
fertilizer_1: [5, 6, 7, 6, 5]
fertilizer_2: [7, 8, 7, 9, 8]
fertilizer_3: [4, 5, 4, 3, 4]


Questions:
- Perform an ANOVA test on the given data. What are the F-value and P-value?
- Based on the P-value, do the fertilizers have significantly different effects on plant growth?
- Explain what would happen if the P-value were greater than 0.05.

```python
from scipy import stats
```

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-value: {f_value}")
print(f"P-value: {p_value}")
```

**What are the F-value and P-value?**

**F-value**: Approximately 28.5 (This is quite high!)

**P-value**: Approximately 0.00002 (Written as 2.7e-05)

**Do the fertilizers have significantly different effects?**


Yes. Because the P-value (0.00002) is much smaller than 0.05, we have strong evidence that the fertilizers do not all work the same way. At least one fertilizer is significantly better (or worse) than the others.

**What if the P-value were greater than 0.05?**


If the P-value were greater than 0.05, we would conclude that there is no statistically significant difference between the fertilizers. In this scenario, any differences in plant growth would be attributed to random chance rather than the effectiveness of the fertilizer itself, it was likely just due to luck (better soil, more sunlight for those specific plants, etc.), and we couldn't prove that one fertilizer is actually better than the others.

Exercise 7: Probability Distributions
===
Work with a binomial distribution: calculate probabilities for different numbers of successes.


Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.

```python
from scipy.stats import binom
```

```python
n = 10    
p = 0.5   
k = 5    

probability = binom.pmf(k, n, p)

print(f"Probability of exactly 5 heads: {probability}")
```

The result of 0.24609375 (or approximately 24.6%) tells us that if we were to flip a coin 10 times, we would get exactly 5 heads about one-quarter of the time.

Even though 5 is the "expected" average, it is not a guarantee. This number proves that while 5 heads is the most common single outcome, there is still a 75.4% chance that you will get something other than 5 (like 4, 6, 3, etc.) due to the natural randomness of the coin.

Exercise 8: Correlation Coefficients
===

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.


Example in code: data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

```python
import pandas as pd
from scipy import stats
```

```python
data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40], 
    'income': [35000, 40000, 50000, 60000, 70000]
})
```

Calculating the Pearson and Spearman:

```python
pearson_coef, p_val_p = stats.pearsonr(data['age'], data['income'])

spearman_coef, p_val_s = stats.spearmanr(data['age'], data['income'])

print(f"Pearson: {pearson_coef}")
print(f"Spearman: {spearman_coef}")
```

The Pearson result of approximately 0.9997 indicates a near perfect positive correlation. While it is not an "exact" 1.0, it shows that age and income move in almost total harmony.
